```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

# Style 
mpl.rcParams['font.family'] = 'DejaVu Sans'
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['legend.fontsize'] = 12

# ROI summary statistics
mean = 5.42
std_dev = 2.50
min_val = 0.75
max_val = 20.77

# Create x values (ROI range in $K)
x = np.linspace(min_val - 3, max_val + 3, 1000)
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev)**2)
cdf = np.cumsum(pdf) / np.sum(pdf)

# Plot
fig, ax = plt.subplots(figsize=(12, 7))
ax.plot(x, cdf, color='#1f77b4', linewidth=3, label="Cumulative Probability")

# Highlight zones
ax.axvline(0, color='black', linestyle='--', linewidth=1)
ax.axvline(5.0, color='green', linestyle='--', linewidth=1)

# Fill areas under the curve
ax.fill_between(x, 0, cdf, where=(x <= 0), color='gray', alpha=0.15, label="Shaded area: ROI ≤ $0 (12.1%)")
ax.fill_between(x, 0, cdf, where=(x >= 0), color='#1f77b4', alpha=0.15, label="Shaded area: ROI > $0 (87.9%)")

# Text annotations
ax.text(-0.4, 0.05, '$0 ROI', ha='right', fontsize=15, color='black')
ax.text(4.9, 0.64, '$50K ROI', ha='center', fontsize=15, color='black')
ax.text(-1.5, 0.98, '12.1%', fontsize=15, ha='center', color='black')
ax.text(8.5, 0.95, '87.9%', fontsize=15, ha='center', color='black')

# Title and axis labels
#ax.set_title("Cumulative Probability Distribution of 3-Year ROI (Model 1)")
ax.set_xlabel("ROI ($ Tens of Thousands)")
ax.set_ylabel("Cumulative Probability")

# Grid and legend
ax.grid(True, linestyle='--', alpha=0.4)
ax.legend(loc='lower right', frameon=False)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

# Style
mpl.rcParams['font.family'] = 'DejaVu Sans'
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['legend.fontsize'] = 12

# Statistics
mean2 = 2.20
std_dev2 = 0.33
min_val2 = 1.31
max_val2 = 3.57

# Create x values
x2 = np.linspace(min_val2 - 1.0, max_val2 + 0.5, 1000)
pdf2 = (1 / (std_dev2 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x2 - mean2) / std_dev2)**2)
cdf2 = np.cumsum(pdf2)
cdf2 = cdf2 / cdf2[-1]  # Normalize

# Plot
fig, ax = plt.subplots(figsize=(12, 7))
ax.plot(x2, cdf2, color='#1f77b4', linewidth=3, label="Cumulative Probability")

# Reference lines
ax.axvline(0, color='black', linestyle='--', linewidth=1)
ax.axvline(1.5, color='green', linestyle='--', linewidth=1)

# Fill areas under the curve
ax.fill_between(x2, 0, cdf2, where=(x2 <= 0), color='gray', alpha=0.15, label="Shaded area: ROI ≤ $0 (12.2%)")
ax.fill_between(x2, 0, cdf2, where=(x2 >= 0), color='#1f77b4', alpha=0.15, label="Shaded area: ROI > $0 (87.8%)")

# Annotations
ax.text(0.05, 0.05, '$0 ROI', ha='left', fontsize=15, color='black')
ax.text(1.48, 0.5, '$15K ROI', ha='center', fontsize=15, color='black')
ax.text(-0.3, 0.95, '12.2%', fontsize=15, ha='center', color='black')
ax.text(2.5, 0.95, '87.8%', fontsize=15, ha='center', color='black')

# Labels and limits
# ax.set_title("Cumulative Probability Distribution of 3-Year ROI (Model 2)")
ax.set_xlabel("ROI ($ Tens of Thousands)")
ax.set_ylabel("Cumulative Probability")
ax.set_xlim(-0.7, 4.2)  # Slightly trimmed left

# Grid and legend
ax.grid(True, linestyle='--', alpha=0.4)
ax.legend(loc='lower right', frameon=False)

plt.tight_layout()
plt.show()
```
